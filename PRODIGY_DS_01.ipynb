```python
df1=data[["Country Code","Region","IncomeGroup","TableName"]]
df1.head(10)
```

```python
df1=df1.copy()
df1.dropna(subset=['Region'], inplace=True)
df1.rename(columns={'TableName' : 'CountryName'}, inplace=True)
df1.head(10)
```

```python
#PLOTTING BARCHART FOR DISTRIBUTION OF REGION
Region_counts = data['Region'].value_counts()
print(f'\ncounts of countries by Region :\n\n{Region_counts}\n')
x=range(len(Region_counts.index))

plt.figure(figsize=(10,5))
plt.bar(Region_counts.index,Region_counts.values,color='#E4080A')
plt.xlabel('Region groups', fontsize=18)
plt.ylabel('Region groups', fontsize=18)
plt.title('Category Distribution of Region',fontsize=18)
plt.xticks(x,Region_counts.index,rotation=45,color='#7DDA58',fontsize=18)
plt.legend(title='Region',fontsize=18)
plt.show()
```

```python
#PLOTTING CHART FOR INCOME GROUP
IncomeGroup_counts = df1['IncomeGroup'].value_counts()
print(f"\nCounts of Countries by IncomeGroups:\n\n{IncomeGroup_counts}\n")
plt.figure(figsize=(10,5))
plt.bar(IncomeGroup_counts.index,IncomeGroup_counts.values,color='#98F5F9')
plt.xlabel('Income Groups', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.title('Category Distribution of IncomeGroup',fontsize=18)
plt.legend(title='IncomeGroup',fontsize=18)
plt.show()

```

```python
#CREATION OF 2D CATEGORIAL DISTRIBUTION OF REGIONAL AND INCOMEGROUP
categorial_distribution=pd.crosstab(df1['Region'], df1['IncomeGroup'])
categorial_distribution.plot(kind='bar', stacked=True, figsize=(10,5), fontsize=14)
plt.xlabel("Region",fontsize=18)
plt.ylabel("count",fontsize=18)
plt.title("2-D categorial Distribution Region vs IncomeGroups",fontsize=18)
plt.tight_layout
plt.show()
```
